````python
# Jupyter Notebook for Python Basics
# Week 1: Getting Started with Machine Learning

## Cell 1: Import Libraries
```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

print("Welcome to the FAAI Course!")
print("Week 1: Python Basics for Machine Learning")
```

## Cell 2: Working with NumPy Arrays
```python
# Create sample data
X = np.array([[1, 2], [3, 4], [5, 6]])
y = np.array([0, 1, 0])

print("Feature matrix X:")
print(X)
print("\nTarget vector y:")
print(y)
```

## Cell 3: Data Visualization
```python
# Create a simple plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Sample Data Visualization')
plt.colorbar()
plt.show()
```

## Cell 4: Basic Pandas Operations
```python
# Create a DataFrame
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
df['Target'] = y

print("Dataset as DataFrame:")
print(df)
print("\nDataset Info:")
print(df.describe())
```

## Exercise
Complete the following tasks:
1. Create your own dataset with 10 samples and 3 features
2. Visualize the data using different plot types
3. Calculate basic statistics (mean, std, min, max)
4. Save your results to a CSV file

## Additional Resources
- NumPy Documentation: https://numpy.org/doc/
- Pandas Documentation: https://pandas.pydata.org/docs/
- Matplotlib Gallery: https://matplotlib.org/gallery/
````
